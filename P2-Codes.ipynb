```python
import pandas as pd

# Define the file path
file_path = 'P2- OfficeSupplies Data.csv'

# Read the CSV data into a DataFrame
sales_data = pd.read_csv(file_path)

# Calculate total sales for each product category
sales_data['Total Sales'] = sales_data['Units'] * sales_data['Unit Price']
category_sales = sales_data.groupby('Item')['Total Sales'].sum().reset_index()

# Identify the product category with the highest sales
top_category = category_sales.loc[category_sales['Total Sales'].idxmax()]

# Identify the top 10 best-selling products
top_products = sales_data.groupby(['Item', 'Rep'])['Units'].sum().reset_index().sort_values(by='Units', ascending=False).head(10)

# Output the results
print("Total Sales for Each Product Category:")
print(category_sales)
print("\nProduct Category with the Highest Sales:")
print(top_category)
print("\nTop 10 Best-Selling Products:")
print(top_products)
```

```python
import pandas as pd

# Define the file path
file_path = 'P2- OfficeSupplies Data.csv'

# Read the CSV data into a DataFrame
sales_data = pd.read_csv(file_path)

# Calculate total sales for each product category
sales_data['Total Sales'] = sales_data['Units'] * sales_data['Unit Price']
category_sales = sales_data.groupby('Item')['Total Sales'].sum().reset_index()

# Identify the product category with the highest sales
top_category = category_sales.loc[category_sales['Total Sales'].idxmax()]

# Identify the top 10 best-selling products
top_products = sales_data.groupby(['Item', 'Rep'])['Units'].sum().reset_index().sort_values(by='Units', ascending=False).head(10)

# Customer Analysis
# - Identify the top 10 customers by sales
# - Calculate the total number of unique customers
# - Analyze customer purchase frequency

# Top 10 customers by sales
top_customers = sales_data.groupby('Rep')['Total Sales'].sum().reset_index().sort_values(by='Total Sales', ascending=False).head(10)

# Total number of unique customers
unique_customers = sales_data['Rep'].nunique()

# Customer purchase frequency
customer_purchase_frequency = sales_data['Rep'].value_counts().reset_index()
customer_purchase_frequency.columns = ['Rep', 'Purchase Frequency']

# Output the results
print("Total Sales for Each Product Category:")
print(category_sales)
print("\nProduct Category with the Highest Sales:")
print(top_category)
print("\nTop 10 Best-Selling Products:")
print(top_products)
print("\nTop 10 Customers by Sales:")
print(top_customers)
print("\nTotal Number of Unique Customers:")
print(unique_customers)
print("\nCustomer Purchase Frequency:")
print(customer_purchase_frequency)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path
file_path = 'P2- OfficeSupplies Data.csv'

# Read the CSV data into a DataFrame
sales_data = pd.read_csv(file_path)

# Convert the OrderDate column to datetime format
sales_data['OrderDate'] = pd.to_datetime(sales_data['OrderDate'], format='%d-%b-%y')

# Calculate total sales for each row
sales_data['Total Sales'] = sales_data['Units'] * sales_data['Unit Price']

# Set the OrderDate as the index
sales_data.set_index('OrderDate', inplace=True)

# Resample the data by month and sum the total sales
monthly_sales = sales_data['Total Sales'].resample('M').sum()

# Plot the monthly sales trends
plt.figure(figsize=(12, 6))
monthly_sales.plot()
plt.title('Monthly Sales Trends')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()

# Output the results
print("Monthly Sales Trends:")
print(monthly_sales)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path
file_path = 'P2- OfficeSupplies Data.csv'

# Read the CSV data into a DataFrame
sales_data = pd.read_csv(file_path)

# Convert the OrderDate column to datetime format
sales_data['OrderDate'] = pd.to_datetime(sales_data['OrderDate'], format='%d-%b-%y')

# Calculate total sales for each row
sales_data['Total Sales'] = sales_data['Units'] * sales_data['Unit Price']

# Sales by Region
region_sales = sales_data.groupby('Region')['Total Sales'].sum().reset_index().sort_values(by='Total Sales', ascending=False)

# Plot total sales by region
plt.figure(figsize=(10, 6))
plt.bar(region_sales['Region'], region_sales['Total Sales'], color=['blue', 'orange', 'green', 'red'])
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()

# Set the OrderDate as the index
sales_data.set_index('OrderDate', inplace=True)

# Resample the data by month and sum the total sales for each region
monthly_region_sales = sales_data.groupby('Region')['Total Sales'].resample('M').sum().unstack('Region')

# Plot the monthly sales trends for each region
plt.figure(figsize=(12, 8))
for region in monthly_region_sales.columns:
    plt.plot(monthly_region_sales.index, monthly_region_sales[region], label=region)

plt.title('Monthly Sales Trends by Region')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.legend(title='Region')
plt.grid(True)
plt.show()

# Output the results
print("Total Sales by Region:")
print(region_sales)
print("\nMonthly Sales Trends by Region:")
print(monthly_region_sales)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path
file_path = 'P2- OfficeSupplies Data.csv'

# Read the CSV data into a DataFrame
sales_data = pd.read_csv(file_path)

# Convert the OrderDate column to datetime format
sales_data['OrderDate'] = pd.to_datetime(sales_data['OrderDate'], format='%d-%b-%y')

# Assume the profit margin is available. For this example, let's add a mock profit margin column.
# Replace this with actual profit data if available.
# Mock profit margin percentage for demonstration
sales_data['Profit Margin'] = 0.2

# Calculate profit for each row
sales_data['Profit'] = sales_data['Units'] * sales_data['Unit Price'] * sales_data['Profit Margin']

# Profit by product category
category_profit = sales_data.groupby('Item')['Profit'].sum().reset_index().sort_values(by='Profit', ascending=False)

# Top 10 most profitable products
top_10_profitable_products = category_profit.head(10)

# Plot total profit by product category
plt.figure(figsize=(10, 6))
plt.bar(category_profit['Item'], category_profit['Profit'], color='purple')
plt.title('Total Profit by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Profit')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()

# Plot top 10 most profitable products
plt.figure(figsize=(10, 6))
plt.bar(top_10_profitable_products['Item'], top_10_profitable_products['Profit'], color='green')
plt.title('Top 10 Most Profitable Products')
plt.xlabel('Product')
plt.ylabel('Total Profit')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()

# Output the results
print("Total Profit by Product Category:")
print(category_profit)
print("\nTop 10 Most Profitable Products:")
print(top_10_profitable_products)
```
